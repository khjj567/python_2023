##### matplotlib 사용

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
arr = np.random.randn(30)
arr
````

````python
# 단일차트
plt.plot(arr)
````

````python
 # 다중차트2 
 fig= plt.figure() #Figure 객체생성
 ax1 = fig.add1 #ckd                                                                            

````

````python
# 차트 / 차트색상, 라인종류, 마커
ig, axes = plt.subplots(2,2)
axes[0][0].hist(np.random.randn(100), bins=20, color='red', alpha=0.5) 
axes[0][1].scatter(np.arange(30), np.random.randn(30))
axes[1][0].plot(np.random.randn(50).cumsum(), linestyle=':', marker='^', color='green', alpha=0.5) 
axes[1][1].plot(np.random.randn(50).cumsum(), linestyle='--', marker='*')
# marker(많다), linestyle(-, --, :, -), color(기본색상(rgb?)))

plt.show()
````

````python
# 범례 legend
plt.plot(np.random.randn(50).cumsum(),label='x')
plt.plot(np.random.randn(50).cumsum(),label='y') 
plt.legend(loc='best') #가장 괜찮은 위치로 범례를 보냄(좌측상단이나 뭐 그런 곳으로) 'upper right' 'lower left' 'lower right' 'upper left' 'right' 'left' 'center' ...

plt.show()
````

##### Pandas 데이터 프레임, 시리즈

````python
# 판다스 시리즈
pd.Series(np.random.randn(1000).cumsum()).plot()
plt.show()
````

````python
# 한글깨짐현상 : 
from matplotlib import font_manager, rc
font_path = 'C:/Windows/Fonts/D2Coding-Ver1.3.2-20180524-all.ttc'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font) # 한글깨짐현상
# matplotlib.rcParams['axes.unicode_minus'] = False # -깨짐 현상

# x,y축 설정
chart = pd.Series(np.random.randn(1000).cumsum()).plot() 
chart.set_xticks([0,250,500,750,1000]) # x축 간격조정
chart.set_yticks([0,25,50]) # y축 간격조정
chart.set_title('한글 가능') # 영어와 숫자는 안깨지는데 한글은 깨짐
chart.set_xticklabels(['zero', 'one', 'two', 'three', 'four'], rotation=-22) # x축을 라벨로 표시 #rotation : 라벨 각도 
chart.grid(True)

plt.show() # 쓸데없는 타입, 결과 메시지 삭제
````

##### 예제연습

캐글 타이타닉 연습문제 csv

주소 : https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv

````python
train = pd.read_csv('https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv')

train.tail()
````

````python
# 성별 생존율 교차도표 생성
gender_survived = pd.crosstab(train['Sex'], train['Survived'])
gender_survived.columns = gender_survived.columns.map({0:"Dead", 1:"Alive"})
````

````python
bar = gender_survived.plot(kind='bar', title='사망자수 / 생존자수')

#차트의 수치표시
for p in bar.patches:
    left, bottom, width, height = p.get_bbox().bounds
    bar.annotate(f'{height:n}',(left+width/2, height*1.02), ha='center')

gender_survived.plot.bar(subplots=True)

plt.show()
````

````python
# 가로 바차트

gender_survived.plot(kind='barh')
gender_survived.plot.barh(subplots=True)

plt.show()
````

````python
# 라인차트

gender_survived.plot(kind='line')
````

````python
# 파이차트
gender_survived['Alive'].plot(kind='pie', title='생존자 비율', autopct='%1.1f%%')

plt.show()
````
