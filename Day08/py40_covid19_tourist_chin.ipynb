```python
# 각 컬럼의 0인 부분이 필터링
# |\ : 다음페이지로 넘어감
condition = (kto_202901['관광']==0) | \
            (kto_202901['상용']==0) | \
            (kto_202901['공용']==0) | \
            (kto_202901['유학/연수']==0) | \
            (kto_202901['기타']==0) 
kto_202901_zero = kto_202901[condition] # 제로데이터를 빼서 따로 담아두기 (전처리의일부)
kto_202901_zero
```
